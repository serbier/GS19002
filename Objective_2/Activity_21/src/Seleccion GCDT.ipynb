```python
import pandas as pd
```

```python
df = pd.read_excel('./../data/repAdap.xlsx')
df = df[df['Adap. M *F4.5']!= 0]
print(df.shape)
df.head()
```

```python
cfield = pd.read_excel('./../data/repAdap.xlsx', sheet_name='cargaCampo')
cfield.head()
```

```python
msDB = pd.read_excel('./../data/Best Heat Selection GCDT Final (1).xlsx', sheet_name='Multispec DB')
msDB['UE'] = msDB['Genotipo']
msDB.head()
```

```python
print(df.shape)
joinDF = df.merge(msDB, on='UE', how='left')
print(joinDF.shape)
```

```python
cDataJoin = joinDF.merge(cfield[['UE','Adap. M *F4.5 F']], on='UE', how='left')
```

```python
selected = list()
fSelected = list()
r = 50

for i,subdf in cDataJoin.groupby(['family', 'f1.2']):
    if (subdf.shape[0]<=6):
        r -= 1
        selected.append(subdf)
    else:
        subdf = subdf.sort_values('Adap. M *F4.5', ascending=True)
        malos = subdf[~pd.isna(subdf['Adap. M *F4.5 F'])]
        malos = malos.sort_values('Adap. M *F4.5 F', ascending=True)     
        if malos.shape[0] > 0:
            sMalos = list()
            counts = malos.groupby('Adap. M *F4.5 F', as_index=False).count()
            if malos[malos['Adap. M *F4.5 F'] == counts.iloc[0]['Adap. M *F4.5 F']].shape[0] <= 1:
                sMalos.append(malos[malos['Adap. M *F4.5 F'] == counts.iloc[0]['Adap. M *F4.5 F']].iloc[0])
                sMalos.append(malos[malos['Adap. M *F4.5 F'] == counts.iloc[1]['Adap. M *F4.5 F']].sort_values('Phi_index',ascending=True).iloc[0])
                sMalos = pd.concat(sMalos, axis=1).T
            else:
                sMalos.append(malos[malos['Adap. M *F4.5 F'] == counts.iloc[0]['Adap. M *F4.5 F']].sort_values('Phi_index',ascending=True).iloc[0:2])
        else:
            sMalos = list()
            sMalos = subdf.sort_values(['Adap. M *F4.5', 'Phi_index'],ascending=True).iloc[:2]
            
        try:
            sMalos = pd.concat(sMalos)
        except TypeError:
    
            pass
        buenos = subdf[~subdf['UE'].isin(sMalos['UE'])]
        buenos = buenos.sort_values(['Adap. M *F4.5','Adap. M *F4.5 F','Phi_index'], ascending=[False,True,False]).iloc[:5]
        tselec = pd.concat([buenos, sMalos])
        fSelected.append(tselec)
        
selected = pd.concat(selected)
fSelected = pd.concat(fSelected)
selected = pd.concat([selected, fSelected])
```

```python
selected.to_csv('./../results/Selection_f45.csv', index=False)
```
